```python
import numpy as np
x=np.linspace(0,5,30)
y=x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y,'r-*')
plt.grid()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
plt.show() # Printing the plot
```

```python
y1=y
y2=x**3
```

```python
plt.plot(x,y1,'r-*',x,y2,'b-o')
plt.grid()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
plt.show() # Printing the plot
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriernted method
```

```python
fig=plt.figure() # Figure object
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.grid()
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # Between 0 and 1

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```
