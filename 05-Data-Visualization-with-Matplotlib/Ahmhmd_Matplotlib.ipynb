```python
import numpy as np
x=np.linspace(0,5,30)
y=x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y,'r-*')
plt.grid()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
plt.show() # Printing the plot
```

```python
y1=y
y2=x**3
```

```python
plt.plot(x,y1,'r-*',x,y2,'b-o')
plt.grid()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
plt.show() # Printing the plot
```
