```python
import numpy as np
x=np.linspace(0,5,30)
y=x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y,'r-*')
plt.grid()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
plt.show() # Printing the plot
```

```python
y1=y
y2=x**3
```

```python
plt.plot(x,y1,'r-*',x,y2,'b-o')
plt.grid()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
plt.show() # Printing the plot
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriernted method
```

```python
fig=plt.figure() # Figure object
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.grid()
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # Between 0 and 1

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)

#axes is an array of plot objects

#for current_ax in axes:
 #   current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Title 1')

axes[1].plot(y,x)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Title 2')


plt.tight_layout()
```

```python
# Figure size and DPI
```

```python
fig=plt.figure(figsize=(8,2),dpi=100) # Size is by inches

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2)) # Size is by inches

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# Legend
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')
#ax.legend(loc=10)
ax.legend(loc=(0.5,0.5))
```

```python
fig =plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',lw=3,alpha=0.5,ls='--',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') # hex code for color
# alpha is for transperancy lw or linewidth 
# ls Line style -- - -. :, Marker, marker size, marker edge width, color: 
```
